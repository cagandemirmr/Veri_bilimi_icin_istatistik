# TEK ÖRNEKLEM  T TESTİ

Örnek ortalamasına ilişkin test yapmak için kullanılır.

```python
import numpy as np
olcumler = np.array([17,160,234,149,145,107,197,75,201,225,211,219,157,145,127,244,163,114,145,65,112,185,202,146,203,224,203,114,188,156,187,154,177,95,165,50,110,216,138,151,166,135,155,84,251,173,131,207,121,120])
```

```python
olcumler[0:10]
```

```python
import scipy.stats as stats
```

```python
stats.describe(olcumler)
```

```python
#Varsayımlar
#Normal Varsayımlar
```

```python
import pandas as pd
```

```python
#Histogram
pd.DataFrame(olcumler).plot.hist()
```

```python
#Qqplot
import pylab
stats.probplot(olcumler,dist="norm",plot=pylab)
pylab.show()
```

```python
#Shapiro–Wilk testi
```

H0:Örnek dağılım ile teorik normal dağılım arasında ist. ol. anl. bir dağılım yoktur.

H1:...fark vardır.

```python
from scipy.stats import shapiro
```

```python
shapiro(olcumler)
```

```python
print("T Testi istatistiği:", str(shapiro(olcumler)[0]))
print("P-value:", str(shapiro(olcumler)[1]))
```

```python
#Hipotez testinin uygulanması
```

```python
stats.ttest_1samp(olcumler,popmean=170)
```

H0: Web sitemizde geçirilen ortalama süre 170'dir.

H1:..değildir.
